# Basics

This notebook introduces foundational Semantic Link Labs tasks: resolving workspace and dataset IDs, listing items, and inspecting model dependencies.

```python
# Parameters
workspace_name = "Demo"
dataset_name = "card_udf"
```

```python
# Imports
# Imports used by demo(s): Listing Items
import sempy.fabric as fab
# Imports used by demo(s): How to find a Workspace and Dataset ID via Semantic Link Labs; Listing all Reports using a Semantic Model; Listing all Datasources a Semantic Model is using; Listing all Tables in a Semantic Model; Dependencies
import sempy_labs as labs
```

## Workspace ID and Dataset ID

##### How to find a Workspace and Dataset ID via the web URL

You can find them when a report is open in the Power BI Web URL

<img src="https://raw.githubusercontent.com/edwardpcharles/presentations/main/Images/url.png" alt="Admin Portal" width="1000" />

##### How to find a Workspace and Dataset ID via Semantic Link Labs

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.resolve_dataset_id

# Get Workspace ID
workspace_id = labs._helper_functions.resolve_workspace_id(workspace_name)

# Get Dataset ID
dataset_id = labs._helper_functions.resolve_dataset_id(dataset=dataset_name, workspace=workspace_name)

# Printing Workspace and Dataset IDs
print(workspace_id)
print(dataset_id)
```

## Listing Items
##### Listing all Semantic Models

```python
# List all datasets
fab.list_datasets(workspace=workspace_name)
```

##### Listing all Reports using a Semantic Model

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.list_reports_using_semantic_model

# List all reports using a semantic model
labs.list_reports_using_semantic_model(
    workspace=workspace_name,
    dataset=dataset_name
)
```

##### Listing all Datasources a Semantic Model is using

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.list_semantic_model_datasources

# List all datasources a semantic model is using
labs.list_semantic_model_datasources(
    dataset=dataset_name,
    workspace=workspace_name
)
```

##### Listing all Tables in a Semantic Model

```python
# List all tables in a semantic model
labs._list_functions.list_tables(
    dataset=dataset_name,
    workspace=workspace_name
)
```

##### Dependencies

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.get_model_calc_dependencies

# List all calc dependencies in a semantic model
labs.get_model_calc_dependencies(
    dataset=dataset_name,
    workspace=workspace_name
)
```
