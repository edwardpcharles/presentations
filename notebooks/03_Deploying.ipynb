# Deploying

This notebook focuses on deployment and promotion workflows, including model movement across workspaces, endorsement, discoverability, and BPA checks.

```python
# Parameters
workspace_name = "Demo"
source_workspace_name = "Demo"
target_workspace_name = "Demo2"
dataset_name = "card_udf"
source_dataset_name = "card_udf"
target_dataset_name = "card_udf"
new_dataset_name = "My New Model"
```

```python
# Imports
# Imports used by demo(s): How to run BPA with custom rules
import sempy.fabric
# Imports used by demo(s): How to run BPA with custom rules
sempy.fabric._client._utils._init_analysis_services()
import Microsoft.AnalysisServices.Tabular as TOM
# Imports used by demo(s): How to run BPA with custom rules
import pandas as pd
# Imports used by demo(s): Semantic Model Movement; How to delete a blank Semantic model; How to Move and Copy a Semantic Model; How to Endorse a Semantic Model; How to Make a Semantic Model Discoverable - Needs to be endorsed first; BPA Analyzer; How to run BPA with custom rules
import sempy_labs as labs
```

## Semantic Model Movement 
##### How to create a blank Semantic model

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.create_blank_semantic_model

# create a empty Semantic Model
labs.create_blank_semantic_model(
    dataset=new_dataset_name,
    workspace=workspace_name
)
```

##### How to delete a blank Semantic model

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.delete_semantic_model

# How to delete a Semantic Model
labs.delete_semantic_model(
    dataset=new_dataset_name,
    workspace=workspace_name
)
```

##### How to Move and Copy a Semantic Model

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.deploy_semantic_model

# Move a Semantic Model from one Workspace to another
labs.deploy_semantic_model(
    source_dataset=source_dataset_name,
    source_workspace=source_workspace_name,
    target_dataset=target_dataset_name,
    target_workspace=target_workspace_name,
    refresh_target_dataset=False,
    overwrite=True
)
```

##### How to Endorse a Semantic Model

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.semantic_model.html#sempy_labs.semantic_model.set_endorsement

# How to Endorse a Semantic Model
# Adds a little badge next to the dataset in Power BI Service to indicate its endorsement status.
labs.semantic_model.set_endorsement(
    dataset=dataset_name,
    workspace=workspace_name,
    endorsement="Promoted"   # "None" | "Promoted" | "Certified" | "Master data"
)
```

##### How to Make a Semantic Model Discoverable - Needs to be endorsed first

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.semantic_model.html#sempy_labs.semantic_model.make_discoverable

# How to Make a Semantic Model Discoverable
labs.semantic_model.make_discoverable(
    dataset=dataset_name,
    workspace=workspace_name,
    make_discoverable=True   # False to remove
)
```

## BPA Analyzer  
##### How to run normally

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.run_model_bpa

#Runs BPA Analyzer with Standard Rules
labs.run_model_bpa(
    dataset=dataset_name,
    workspace=workspace_name
)
```

##### How to run BPA with custom rules

```python
# Semantic Link Labs docs reference
# - https://semantic-link-labs.readthedocs.io/en/latest/sempy_labs.html#sempy_labs.run_model_bpa

custom_rules = pd.DataFrame([
    (
        "Formatting",           # Category — groups rules in the output
        "Measure",              # Scope — TOM object type to check
        "Warning",              # Severity: "Warning" | "Error" | "Info"
        "Measures must have a format string",   # Rule Name
        lambda obj, tom: not obj.FormatString,  # Expression — True = violation
        "All measures should have a format string defined.",  # Description
        None,                   # URL — link to docs (optional)
    ),
    (
        "Organization",
        "Measure",
        "Warning",
        "Measures should have a display folder",
        lambda obj, tom: not obj.DisplayFolder,
        "All measures should be organized into display folders.",
        None,
    ),
],
columns=["Category", "Scope", "Severity", "Rule Name", "Expression", "Description", "URL"])

labs.run_model_bpa(
    dataset=dataset_name,
    workspace=workspace_name,
    rules=custom_rules
)
```
