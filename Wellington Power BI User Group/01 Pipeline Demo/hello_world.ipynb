```python
import json

# Default expected value if no parameter is passed
expected_output = "Hello World"

# Try to read Fabric notebook parameter
try:
    expected_output = notebookutils.widgets.get("ExpectedOutput")
except Exception:
    pass

actual_output = "Hello World"
is_match = (actual_output == expected_output)

result = {
    "expected_output": expected_output,
    "actual_output": actual_output,
    "result": is_match
}

print(json.dumps(result, indent=2))

# Fail the run when outputs do not match so pipeline can fail reliably.
if not is_match:
    raise ValueError(
        f"Output mismatch. expected={expected_output}, actual={actual_output}"
    )

# Return notebook output as true/false for pipeline consumers
try:
    notebookutils.notebook.exit(str(is_match).lower())
except Exception:
    pass
```
